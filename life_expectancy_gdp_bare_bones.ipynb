```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Examine the data file
df = pd.read_csv('all_data.csv')
print(df.head())
df['billions_gdp'] = df['GDP']/1000000000
print(df.head()) 
print(df.Country.unique())
print(df.Year.unique())
```

```python
# My goals are to use Matplotlib and Seaborn to answer these questions:
# 1. What is the relationship between life expectancy and GDP?
# 2. How has life expectancy changed over the years?
# 3. How has GDP changed over the years?
# 4. What other questions would merit further study?
print('A portfolio project by William Malpass')
```

```python
df.columns = ['country', 'year', 'life_expectancy', 'exp_gdp', 'gdp']
print(df.head())
```

```python
chile = df[df.country == 'Chile']
china = df[df.country == 'China']
germany = df[df.country == 'Germany']
mexico = df[df.country == 'Mexico']
usa = df[df.country == 'United States of America']
zimbabwe = df[df.country == 'Zimbabwe']
years = df.year.unique()
```

```python
for x in [chile, china, germany, mexico, usa, zimbabwe]:
    plt.figure(figsize=(4,8))
    sns.lmplot(x='gdp', y='life_expectancy', data = x)
    plt.title('GDP vs. Life Expectancy for ' + str(x.country.iloc[0]))
    plt.xlabel('GDP (Billions of USD)')
    plt.ylabel('Life Expectancy (Years)')
    plt.subplots_adjust(top=0.8)
    plt.savefig(str(x.country.iloc[0]) + '.png')
    plt.show()
    plt.clf()
```

```python
plt.plot(years, chile.life_expectancy, marker='o')
plt.plot(years, china.life_expectancy, marker='o')
plt.plot(years, germany.life_expectancy, marker='o')
plt.plot(years, mexico.life_expectancy, marker='o')
plt.plot(years, usa.life_expectancy, marker='o')
#plt.plot(years, zimbabwe.life_expectancy, marker='o')

plt.title('Life Expectancy (2000-2015)')
plt.ylabel('Life Expectancy (Years)')
plt.legend(['Chile', 'China', 'Germany', 'Mexico', 'USA', 'Zimbabwe'])
plt.savefig('first_five.png')
plt.show()
plt.clf()
```

```python
plt.plot(years, zimbabwe.life_expectancy, marker='o')

plt.title('Life Expectancy (2000-2015)')
plt.ylabel('Life Expectancy (Years)')
plt.legend(['Zimbabwe'])
plt.savefig('lastly_zimbabwe.png')
plt.show()
plt.clf()
```
